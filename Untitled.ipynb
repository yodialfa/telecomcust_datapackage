```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
#from chromedriver_py import binary_path
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import time
```

```python

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',options=chrome_options)
```

```python

link = "https://www.bukalapak.com/products?search%5Bkeywords%5D=paket%20data"
driver.get(link)
res = driver.page_source.encode("utf-8")
soup = BeautifulSoup(res, "html.parser")
time.sleep(5)
driver.quit()
```

```python
#driver.set_windows_size(1300,800)
#driver.save_screenshot('home.png')
#content = driver.page_source
# scroll = 500
# for i in range (1,7):
#     akhir = scroll * 1
#     perintah = "window.scrollTo(0,"+str(akhir)+")"
#     driver.execute_script(scroll)
#     time.sleep(1)


#while True:
    #WebDriverWait(driver,5 ).until(ec.presence_of_element_located((By.CSS_SELECTOR,"#product-explorer-container > div > div.bl-flex-container > div.bl-flex-item.bl-product-list-wrapper > div > div:nth-child(2) > nav > ul > li:nth-child(2)"))).click()
    #time.sleep(2) 
#product-explorer-container > div > div.bl-flex-container > div.bl-flex-item.bl-product-list-wrapper > div > div:nth-child(2) > nav > ul > li:nth-child(2) > a
#create column
            
product_name, product_price, product_rating, product_city =[],[],[],[]
i=1
for area in soup.find_all('div',class_="bl-product-card te-product-card"):
    nama = area.find('a',class_="bl-link").get_text()
    #print("iterasi nama",i)
    harga = area.find('div', class_="bl-product-card__description-price").get_text()
    #print("iterasi harga",i)
    # terjual = area.find('p', class_="bl-text bl-text--body-14 bl-text--subdued").get_text()
    rating = area.find('div', class_="bl-product-card__description-rating-and-sold").get_text()
    #print("iterasi rating",i)
    kota = area.find('span', class_="mr-4 bl-product-card__location bl-text bl-text--body-14 bl-text--subdued bl-text--ellipsis__1").get_text()
    #print("iterasi kota",i)


    product_name.append(nama)
    product_price.append(harga)
    product_rating.append(rating)
    product_city.append(kota)
    i+=1
    # print(i)
    
```
